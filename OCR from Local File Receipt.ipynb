```python
from PIL import Image
from paddleocr import PaddleOCR,draw_ocr
import os
import glob
import pandas as pd
```

```python
#Read Picture, Than Save Text Results 
folder_dir = "D:\\content\\test"
resc=[]
for image in os.listdir(folder_dir):
    file = glob.glob('D:/content/test/' + image) 
    for filee in file:
        result = ocr_model.ocr(filee, cls=True)
        filename=image
        resc.append([filename, result])
    df_res=pd.DataFrame(resc, columns=['File', 'Result'])
    df_res
```

```python
#Merge File Name with Results
df_res_n=df_res.copy()
list_res=[]
appendtext=[]
seen = set(appendtext)
df_fix=pd.DataFrame()
alldata=len(df_res_n)
df_compile=pd.DataFrame()

for num in range (0,alldata):
    for res in df_res_n['Result'][num]:
        for i in range(len(res)):
            textout=res[i][1][0]
            num1=num
            if (num, textout) not in seen:
                seen.add(num)
                seen.add(textout)
                appendtext.append([num, textout])

    df_textcom=pd.DataFrame(appendtext, columns=['file_number', 'Extracted_Text'])
    df_textcom=df_textcom.drop_duplicates()
    df_fix=df_textcom.groupby('file_number', as_index=False).agg(lambda x: x.tolist())
    dropnumber=df_fix.reset_index()
    dropnumber=dropnumber.drop(['file_number', 'index'], axis=1)
    df_compile=dropnumber.join(df_withname)
    
```
